```python
class Nodo:
    def __init__(self, valor=None, hijos=None, result=None):
        self.valor = valor
        self.hijos = hijos or []
        self.result = result or []


def generar_arbol_expresiones(numeros, operadores):
    raiz = Nodo()
    for i, num1 in enumerate(numeros):
        for j, op1 in enumerate(operadores):
            for k, num2 in enumerate(numeros):
                if i != k:
                    for l, op2 in enumerate(operadores):
                        if j != l:
                            for m, num3 in enumerate(numeros):
                                if i != m and k != m:
                                    for n, op3 in enumerate(operadores):
                                        if j != n and l != n:
                                            for o, num4 in enumerate(numeros):
                                                if i != o and k != o and m != o:
                                                    for p, op4 in enumerate(operadores):
                                                        if j != p and l != p and n != p and o != p:
                                                            for q, num5 in enumerate(numeros):
                                                                if i != q and k != q and m != q and o != q  :
                                                                    nodo = Nodo([num1, op1, num2, op2, num3, op3, num4, op4, num5])
                                                                    raiz.hijos.append(nodo)
                                                                    raiz.result.append(''.join(str(e) for e in [num1, op1, num2, op2, num3, op3, num4, op4, num5]))
    return raiz


def imprimir_arbol(nodo):
    if nodo.valor is not None and len(nodo.valor) == 9:
        print(nodo.valor)
    for hijo in nodo.hijos:
        imprimir_arbol(hijo)


numeros = [1,2,3,4,5,6,7,8,9]
operadores = ['+', '-', '*', '/']

arbol_expresiones = generar_arbol_expresiones(numeros, operadores)

#imprimir_arbol(arbol_expresiones)
```

```python
def buscar_combinacion_entera(nodo, resultado):
    if nodo.valor is not None and len(nodo.valor) == 9:
        expresion = ''.join(map(str, nodo.valor))
        try:
            valor = eval(expresion)
            if valor.is_integer() and int(valor) == resultado:
                print(expresion)
        except:
            pass
    for hijo in nodo.hijos:
        buscar_combinacion_entera(hijo, resultado)
```

```python
def buscar_minimo_entero(nodo):
    if nodo.valor is not None and len(nodo.valor) == 9:
        expresion = ''.join(map(str, nodo.valor))
        try:
            valor = eval(expresion)
            if valor.is_integer() and (not hasattr(buscar_minimo_entero, 'minimo') or valor < buscar_minimo_entero.minimo):
                buscar_minimo_entero.minimo = valor
        except:
            pass
    for hijo in nodo.hijos:
        buscar_minimo_entero(hijo)
```

```python
def buscar_maximo_entero(nodo):
    if nodo.valor is not None and len(nodo.valor) == 9:
        expresion = ''.join(map(str, nodo.valor))
        try:
            valor = eval(expresion)
            if valor.is_integer() and (not hasattr(buscar_maximo_entero, 'maximo') or valor > buscar_maximo_entero.maximo):
                buscar_maximo_entero.maximo = valor
        except:
            pass
    for hijo in nodo.hijos:
        buscar_maximo_entero(hijo)
```

```python
def comprobar_valores_en_arbol(raiz, valor_minimo, valor_maximo):
    valores_en_arbol = set()

    # Recorrer el árbol en anchura
    nodos_por_nivel = [[raiz]]
    for nivel in nodos_por_nivel:
        siguiente_nivel = []
        for nodo in nivel:
            # Procesar el nodo actual
            if nodo.valor is not None and len(nodo.valor) == 9:
                expresion = ''.join(map(str, nodo.valor))
                try:
                    valor = eval(expresion)
                    if valor.is_integer():
                        valores_en_arbol.add(valor)
                except:
                    pass
            # Agregar los hijos del nodo actual al siguiente nivel
            for hijo in nodo.hijos:
                siguiente_nivel.append(hijo)
        if siguiente_nivel:
            nodos_por_nivel.append(siguiente_nivel)

    # Verificar si todos los valores desde valor_minimo hasta valor_maximo están en el conjunto de valores del árbol
    for i in range(valor_minimo, valor_maximo + 1):
        if i not in valores_en_arbol:
            return False
    return True
```
